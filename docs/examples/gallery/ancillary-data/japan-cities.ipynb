## Cities in Japan

```python
import earthkit.maps

chart = earthkit.maps.Chart(domain="Japan")

chart.land()
chart.urban_areas()
chart.coastlines()
chart.rivers()
chart.lakes()
chart.borders()
chart.states_provinces()
chart.ocean()
chart.gridlines(draw_labels=False)
chart.cities(density="high")
    
chart.show()
```
