## Built-in domain names

```python
import earthkit.maps

chart = earthkit.maps.Chart(rows=2, columns=3)

for domain in [
        "Brazil",
        "North Atlantic",
        "Antarctica",
        "Africa",
        "Greenland",
        "Australia",
]:
    chart.add_subplot(domain=domain)

chart.stock_img()
chart.coastlines()
chart.borders()
chart.gridlines()

chart.subplot_titles("{domain}")

chart.show()
```
