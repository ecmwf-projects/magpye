```python
import earthkit.maps
```

```python
chart = earthkit.maps.Chart(domain="Greenland")
chart.coastlines()
chart.gridlines()
chart.show()
```

You might want the coastlines to be blue, or the gridlines to be thicker, solid lines. You can of course achieve this by passing the relevant arguments straight into the `Chart` methods:

```python
chart = earthkit.maps.Chart(domain="Greenland")
chart.coastlines(color="blue")
chart.gridlines(linestyle="solid", linewidth=4)
chart.show()
```

But what if you *always* want this style? This is where the `earthkit.schema` comes in:

```python
earthkit.maps.schema.coastlines.set(color="blue")
earthkit.maps.schema.gridlines.set(linestyle="solid", linewidth=4)

chart = earthkit.maps.Chart(domain="Greenland")
chart.coastlines()
chart.gridlines()
chart.show()
```

You can also temporarily set parameters using the `schema` as a context manager:

```python
chart = earthkit.maps.Chart(domain="Greenland")
chart.coastlines()

with earthkit.maps.schema.gridlines.set(linestyle=":"):
    chart.gridlines()

chart.show()
```
