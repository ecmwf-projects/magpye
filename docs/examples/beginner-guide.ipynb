## Beginner's Guide

Welcome to `earthkit.maps`!

`earthkit.maps` speeds up visualisation of meterological and climate data by providing high-level interfaces which reduce boilerplate code.

As part of the `earthkit` ecosystem, `earthkit.maps` is built on top of the powerful data interfaces provided by `earthkit.data`. The best way to get started with producing some plots is to open some data with `earthkit.data`.

```python
import earthkit.data
```

```python
data = earthkit.data.from_source("file", "era5.grib")
data.ls()
```

Now that we have some data, let's instantiate a `Chart` object.

```python
from earthkit.maps import Chart
```

```python
chart = Chart()

chart.shaded_contour(data)

chart.borders()
chart.coastlines()
chart.gridlines()
chart.show()
```

Charts produced with `earhkit.maps` are **data aware** - that is to say, metadata associated with the data you plot is retained by the chart to greatly improve convenience and automation for elements like map projections, titles, legends and colour palettes.

We can add a `title` and a `legend` to the above plot using some nice defaults based on the underlying data.

```python
chart = Chart()

chart.contourf(data)
chart.title()
chart.legend()

chart.borders()
chart.coastlines()
chart.gridlines()
chart.show()
```
