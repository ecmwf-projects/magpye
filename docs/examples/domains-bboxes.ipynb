## Domains, projections and bounding boxes

```python
import earthkit
from earthkit.maps import Chart
```

```python
data = earthkit.data.from_source("file", "era5.grib")
```

```python
chart = Chart(domain="Norway")
chart.shaded_contour(data)
chart.coastlines()
chart.borders()
chart.gridlines()
chart.legend()
chart.show()
```

```python
import cartopy.crs as ccrs

chart = Chart(domain="Norway", crs=ccrs.PlateCarree())
chart.shaded_contour(data)
chart.coastlines()
chart.borders()
chart.gridlines()
chart.legend()

chart.show()
```

```python
EUROPE_BBOX = (-20, 40, 34, 72)

chart = Chart(domain=EUROPE_BBOX)
chart.shaded_contour(data)
chart.coastlines()
chart.borders()
chart.gridlines()
chart.subplot_titles("{domain}")
chart.show()
```

### Subplots on different domains

```python
earthkit.maps.schema.set(cmap="turbo")

chart = Chart(nrows=1, ncols=3, wspace=0.4)

chart.add_subplot(domain="France")
chart.add_subplot(domain="Italy")
chart.add_subplot(domain="Spain")

chart.shaded_contour(data)
chart.coastlines()
chart.gridlines()
chart.borders()

chart.subplot_titles("{domain}")

chart.show()
```
